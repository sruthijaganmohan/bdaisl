```python
# correlation between variables
plt.figure(figsize=(7, 6))
sns.heatmap(df.corr(), annot=True)
```

```python
# statewise distribution of active cases
colors = ['#4B2991', '#952EA2', '#D44292', '#F66D7A', '#F6A97A'] 
plt.figure(figsize=(10, 8))
plt.pie(df['Total Cases'], labels=df['State/UTs'], colors=colors)
plt.show()
```

```python
# statewise distribution of active cases 
plt.figure(figsize=(10, 8))
sns.barplot(data=df, x='Active', y='State/UTs', palette='magma')
```

```python
# statewise distribution of deceased cases
plt.figure(figsize=(10, 8))
sns.barplot(data=df, x='Deaths', y='State/UTs', palette='magma')
```

```python
# statewise distribution of discharged/recovered cases
plt.figure(figsize=(10, 8))
sns.barplot(data=df, x='Discharged', y='State/UTs', palette='magma')
```
